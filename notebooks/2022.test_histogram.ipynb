```python
from hypothesis import given, strategies as hy
@given(
    vals=hy.lists(hy.integers(min_value=1, max_value=1000)),
    cts=hy.lists(hy.integers(min_value=1, max_value=1000)),
)
def fun(vals, cts):
    try:
        pu.FatigueStress(cts, vals, bin_width=0.5)
    except:
        pass
    print('vals', vals)
    print('cts', cts)
```

```python
@hy.composite
def same_len_lists(draw):

    n = draw(hy.integers(min_value=2, max_value=50))
    fixed_length_list = hy.lists(hy.integers(), min_size=n, max_size=n)

    return (draw(fixed_length_list), draw(fixed_length_list))
```

```python
@given(same_len_lists())
def test_my_func(lists):

    x, y = lists
    print(x, y[1:])
    # Assertions
```

```python
import unittest
from typing import Tuple, List
class TestFatigueStress(unittest.TestCase):
    """Test the FatigueStress class"""

    # def setUp(self):
    #     """
    #     Set up for some of the tests.
    #     """

    @given(same_len_lists())
    def test_faulty_fatigue_stress(self, lists: Tuple[List[int], List[int]]):
        """Assert that an error is raised when the counts and values
        have different lengths
        """
        vals, cts = lists
        with self.assertRaises(
            ValueError, "counts and values must have the same length"
        ) as ve:
            _ = pu.FatigueStress(_counts=cts, _values=vals[1:], bin_width=0.5)
```
